# Project Title: 
        Customer Purchase Analysis
# Objective:
        Analyze customer demographics, spending behavior, and income distribution.

        Identify trends in purchases across different product categories and cities.

        Visualize insights using Python & Power BI.

```python
import pandas as pd
import numpy as np

```

```python
np.random.seed(42)
```

```python
num_customers = 1000
```

```python
customer_data = pd.DataFrame({
    "Customer_ID": np.arange(1, num_customers + 1),
    "Age": np.random.randint(18, 65, num_customers),
    "Gender": np.random.choice(["Male", "Female"], num_customers),
    "Annual_Income": np.random.randint(20000, 120000, num_customers),
    "Spending_Score": np.random.randint(1, 100, num_customers),
    "Purchase_Amount": np.random.randint(100, 5000, num_customers),
    "Purchase_Category": np.random.choice(["Electronics", "Clothing", "Groceries", "Furniture"], num_customers),
    "City": np.random.choice(["New York", "Los Angeles", "Chicago", "Houston", "Phoenix"], num_customers)
})
```

# Load & Clean the Data

```python
customer_data.to_csv("customer_purchase_analysis.csv", index=False)
```

# Exploratory Data Analysis (EDA)
Use Pandas, Matplotlib & Seaborn to analyze customer trends.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(df["Age"], bins=20, kde=True, color='blue')
plt.title("Age Distribution of Customers")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(x="Annual_Income", y="Spending_Score", hue="Gender", data=df)
plt.title("Spending Score vs. Annual Income")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x="Purchase_Category", data=df, palette="viridis")
plt.title("Purchase Category Distribution")
plt.show()
```
